### Question (c)
There are 5 different media types.

```python
q_d = """
SELECT
    mt.MediaTypeId,
    mt.Name AS media_type,
    COUNT(t.TrackId) AS track_count
FROM media_types mt
LEFT JOIN tracks t
    ON t.MediaTypeId = mt.MediaTypeId
GROUP BY mt.MediaTypeId, mt.Name
ORDER BY track_count ASC
"""
with sqlite3.connect('data/chinook.db') as con:
    d_df = pd.read_sql(q_d, con)
d_df
```

### Question (d)
Yes, all media types are represented in the tracks table.

```python
q_e = """
SELECT
    mt.Name AS media_type,
    COUNT(t.TrackId) AS track_count
FROM tracks t
JOIN media_types mt
    ON t.MediaTypeId = mt.MediaTypeId
GROUP BY mt.Name
ORDER BY track_count DESC
"""
with sqlite3.connect('data/chinook.db') as con:
    e_df = pd.read_sql(q_e, con)
e_df
```

```python
q_f = """
SELECT
    City,
    COUNT(*) AS customer_count
FROM customers
GROUP BY City
ORDER BY customer_count DESC
LIMIT 1
"""
with sqlite3.connect('data/chinook.db') as con:
    f_df = pd.read_sql(q_f, con)
f_df
```

### Question (f)
The city with the largest number of customers is Sao Paulo.

```python
q_g = """
SELECT
    g.Name AS genre,
    COUNT(t.TrackId) AS track_count
FROM tracks t
JOIN genres g
    ON t.GenreId = g.GenreId
GROUP BY g.Name
ORDER BY track_count DESC
LIMIT 1
"""
with sqlite3.connect('data/chinook.db') as con:
    g_df = pd.read_sql(q_g, con)
g_df
```
